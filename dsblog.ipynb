```python
import pandas as pd 
import numpy as np
import random
import matplotlib.pyplot as plt
from math import pi
import seaborn as sns
```

```python
random.seed(42)
```

```python
#Defining the regions           
regions=["North Kerala","South Kerala","Central Kerala","Coastal Kerala","Hill Kerala"]
```

```python
data={
"Region": regions,
    "Quantum Governance Index": [round(random.uniform(80, 90), 1) for _ in regions],
    "AI-Enhanced Policy Effectiveness": [round(random.uniform(85, 95), 1) for _ in regions],
    "Virtual Town Hall Participation": [f"{random.randint(70, 90)}%" for _ in regions],
    "Neural Public Sentiment Analysis": [f"{random.randint(80, 95)}%" for _ in regions]
}
```

```python
df = pd.DataFrame(data)
csv_file_path = '/home/devan/Blog/political_data.csv'
df.to_csv(csv_file_path, index=False)
print(f"Dataset exported to {csv_file_path}")
```

```python
def plot_radar_chart(data, categories, title):
    plt.figure(figsize=(8, 8))
    
    # Number of variables
    num_vars = len(categories)
    
    # Compute angle of each axis
    angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
    angles += angles[:1]
    
    # Initialize the radar plot
    ax = plt.subplot(111, polar=True)
    
    # Draw one axe per variable + add labels
    plt.xticks(angles[:-1], categories, color='grey', size=10)
    
    # Plot data
    for i, region in enumerate(data['Region']):
        values = data.iloc[i, 1:].tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=region)
        ax.fill(angles, values, alpha=0.4)
    
    # Add a title
    plt.title(title, size=14, color='black', y=1.1)
    
    # Add a legend
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Select relevant columns for the radar chart
radar_data = df[['Region', 'Quantum Governance Index', 'AI-Enhanced Policy Effectiveness', 'Virtual Town Hall Participation', 'Neural Public Sentiment Analysis']]

# Plot the radar chart
plot_radar_chart(radar_data, categories=['Quantum Governance', 'AI Policy Effectiveness', 'Virtual Town Hall', 'Neural Sentiment'], title='Political Decision-Making Metrics by Region')
plt.show()
```

```python
pairplot_data = df[['Quantum Governance Index', 'AI-Enhanced Policy Effectiveness', 'Virtual Town Hall Participation', 'Neural Public Sentiment Analysis']]

# Calculate the correlation matrix
correlation_matrix = pairplot_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Correlation Heatmap of Political Decision-Making Metrics')
plt.show()
```
